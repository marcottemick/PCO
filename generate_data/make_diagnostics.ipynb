```python
import pandas as pd
```

```python
dtypes = {
    'id_diag':int,
    'diagnostic': str,
    'generation': int,
    'statut': str
}
diagnostics = pd.read_csv('../data/diagnostics_v2.csv', sep=';', dtype=dtypes)
diagnostics.head(5)
```

```python
diagnostics = diagnostics[['diagnostic', 'generation', 'statut']]
# diagnostics = diagnostics.sort_values(by='diagnostic', ignore_index=True)
diagnostics
```

```python
diagnostics_new = ["prurigo nodulaire", "eczéma atopique",  "polyangéite granulomateuse", "syphilis", "sarcoïdose", "polyarthrite psoriasique"]

for diag in diagnostics_new:
    new_diagnostic = diag.strip()\
    # .lower()
    if new_diagnostic != "" and new_diagnostic not in diagnostics.diagnostic.to_list():
        diagnostics.loc[len(diagnostics)] = {"diagnostic": new_diagnostic}
        # diagnostics = diagnostics.sort_values(by='diagnostic', ignore_index=True)
    diagnostics
```

```python
diagnostics['id_diag'] = diagnostics.index
diagnostics['generation'] = diagnostics['generation'].fillna(0)
diagnostics['statut'] = diagnostics['statut'].fillna('Train')
diagnostics = diagnostics[['id_diag', 'diagnostic', 'generation', 'statut']]
diagnostics.id_diag = diagnostics.id_diag.astype(int)
diagnostics.generation = diagnostics.generation.astype(int)

diagnostics.to_csv('../data/diagnostics_v2.csv', index=False, sep=';')
```
