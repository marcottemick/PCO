```python
import pandas as pd
import re
import random

patients_df = pd.read_csv('../data/patients.csv', sep=';', dtype=str)
patients_df = pd.concat([patients_df[0: 10], patients_df[200: 210]], ignore_index=True)
meds_df = pd.read_csv('../data/anatomopathologists.csv', sep=';', dtype=str)
meds_df = pd.concat([meds_df[0: 10], meds_df[50: 60]], ignore_index=True)

patients = pd.read_csv('../data/patients.csv', sep=';', dtype=str)
patients = pd.concat([patients[0: 10], patients[200: 210]], ignore_index=True)
meds = pd.read_csv('../data/anatomopathologists.csv', sep=';', dtype=str)
meds = pd.concat([meds[0: 10], meds[50: 60]], ignore_index=True)
diag = pd.read_csv('../data/diagnostics.csv', sep=';')
```

```python
text = """
 Résumé anatomopathologique

Patiente : Mme Laure Caron
Date de naissance : 17/11/1951
Adresse : 912, rue Henry, 76743 Saint Renéenec
Numéro de sécurité social : 2 1951 11 76743 557 98

I. Présentation des données personnelles
Une femme de 70 ans a été renvoyée au laboratoire d'anatomie pathologique pour une étude historique d'un échantillon de tissue issu d'une excision chirurgicale localisée sur la peau.

II. Description macrogamique
L'échantillon reçu avait approximativement une taille de 2 cm x 1,5 cm x 1 cm. Sa coule était brun sombre, avec présence d'un nodule bien limité et homogène. Aucune nécrosis apparente ou signe inflammatoires n'était observable.

III. Description microscopique
À l'examen histologique, l'échantillon montrait une prolifération tumorale dense formant des cordons et des nids irréguliers. Les cellules fusiformes présentaient un noyau ovale, hyperchromatique et irrégulier, ainsi que des mitoses atypiques abondantes. Des espaces intercellulaires sanguineux et des vaisseaux sanguins dilatés étaient aussi visibles. Par immunohistochimie, les cellules ont exprimé une positivité pour HHV-8 et CD30, confirmant le diagnose de Sarcome à cells fusiformes de Kaposi (SCFK).

IV. Conclusion
D'après cette étude histologique, le diagnostic final est celui d'un sarcome à cells fusiformes de Kaposi. Les aspects morphologique et immunohistochimique sont concordants avec ce diagnostic. Il est important de communiquer ces résultats aux praticiens soignants et de tenir compte d'une surveillance régulière afin d'éviter tout risque de récidive ou de métastases possibles.

Dr Inès Cohen
Anatomopathologiste

Note : Ce document est un compte rendu fictif."""

diag_search = 'sarcome à cellules fusiformes de Kaposi'

text = text.replace('Docteur', 'Dr')

datas_diag = diag.query(f'diagnostic == "{diag_search}"')
print(datas_diag)
datas_diag = datas_diag.to_dict(orient= 'records')

regex_nir = r'\d+ \d+ \d+ \d+ \d+ \d+'
nir_in_text  = re.search(regex_nir, text).group(0)

regex_doc = r'Dr\s+\w+\s+\w+'
doc_in_text = re.search(regex_doc, text).group(0)

patient_in_text = patients_df.query(f'nir == "{nir_in_text}"')
datas_patient_in_text = patient_in_text.to_dict(orient='records')

doc_in_text = meds_df.query(f'name == "{doc_in_text}"')
datas_doc_in_text = doc_in_text.to_dict(orient= 'records')

print(len(patients))
print(len(meds))

random_patient = random.randint(0, len(patients) - 1)
random_doc = random.randint(0, len(meds) - 1)

patient_select = patients.loc[random_patient]
med_select = meds.loc[random_doc]

patients = patients.drop(random_patient)
patients = patients.reset_index(drop=True)
meds = meds.drop(random_doc)
meds = meds.reset_index(drop=True)

print(len(patients))
print(len(meds))

if len(patients) == 0:
    patients = pd.read_csv('../data/patients.csv', sep=';', dtype=str)
    patients = pd.concat([patients[0: 10], patients[200: 210]], ignore_index=True)

if len(meds) == 0:
    meds = pd.read_csv('../data/anatomopathologists.csv', sep=';', dtype=str)
    meds = pd.concat([meds[0: 10], meds[50: 60]], ignore_index=True)

patient_select = patient_select.to_dict()
med_select = med_select.to_dict()

print(patient_select)
print(med_select)
print(datas_diag)
print(datas_doc_in_text)
print(datas_patient_in_text)

text_mod = text.replace(datas_doc_in_text[0]['name'], med_select['name'])\
               .replace(datas_patient_in_text[0]['name'], patient_select['name'])\
               .replace(datas_patient_in_text[0]['birthday'], patient_select['birthday'])\
               .replace(datas_patient_in_text[0]['address'], patient_select['address'])\
               .replace(datas_patient_in_text[0]['nir'], patient_select['nir'])

new_CRO = pd.DataFrame({
    'CRO': [text_mod],
    'id_diag':[datas_diag[0]['id_diag']],
    'id_med': [med_select['id_med']],
    'nir': [patient_select['nir']],
    'source': 'mistralai/Mistral-7B-Instruct-v0.2',
    'operation' :'exérèse'
})

CROs = pd.read_csv(r'G:\Mon Drive\PCO\data\CRO_df.csv', sep=';')
CROs = pd.concat([CROs, new_CRO], ignore_index=True)
CROs.to_csv(r'G:\Mon Drive\PCO\data\CRO_df.csv', index=False, sep=';')
```
