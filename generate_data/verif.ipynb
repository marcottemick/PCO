```python
import pandas as pd
file = 'G:/Mon Drive/PCO/data/'
```

```python
dtypes = {
    'CRO':str,
    'id_med' : int,
    'id_diag': int,
    'source':str,
    'nir':str,
    'operation':str
}

CRO_1 = pd.read_csv(f'{file}\CRO_1.csv', sep=';', dtype=str)
print('nb de CRO_1 total:', len(CRO_1))
CRO_2_1 = pd.read_csv(f'{file}\CRO_2_1.csv', sep=';', dtype=str)
print('nb de CRO_2_1 total:', len(CRO_2_1))
CRO_2_2 = pd.read_csv(f'{file}\CRO_2_2.csv', sep=';', dtype=str)
print('nb de CRO_2_2 total:', len(CRO_2_2))
CRO_3 = pd.read_csv(f'{file}\CRO_3.csv', sep=';', dtype=dtypes)
print('nb de CRO_3 total:', len(CRO_3))
CRO_4 = pd.read_csv(f'{file}\CRO_4.csv', sep=';', dtype=dtypes)
print('nb de CRO_4 total:', len(CRO_4))
CRO_5 = pd.read_csv(f'{file}\CRO_5.csv', sep=';', dtype=str)
print('nb de CRO_5 total:', len(CRO_5))
print(len(CRO_1) + len(CRO_2_1) + len(CRO_2_2) + len(CRO_3) + len(CRO_4) + len(CRO_5))
```

```python
CRO_1_et_2 = pd.concat([CRO_1, CRO_2_1, CRO_2_2], ignore_index=True)
len(CRO_1_et_2)
```

```python
diags_correspondance = pd.read_csv(f'../data/old_id_diag.csv', sep=';', dtype=str)
diags_correspondance
```

```python
CRO_1_et_2 = pd.merge(CRO_1_et_2, diags_correspondance, left_on="id_diag", right_on='id_diag_old', how='left')
CRO_1_et_2.head(5)
```

```python
CRO_1_et_2 = CRO_1_et_2.drop(['TNM', 'id_diag_x', 'id_diag_old'], axis=1)
CRO_1_et_2 = CRO_1_et_2.rename(columns = {"id_diag_y" : 'id_diag'})
CRO_1_et_2
```

```python
CRO_1_et_2.to_csv(f'{file}/CRO_1_et_2.csv', index=False, sep=';')
```

```python
new_diags = pd.read_csv('../data/diagnostics_v2.csv', sep=';', dtype=str)
diags_copy = pd.read_csv('../data/diagnostics copy.csv', sep=';', dtype=str)
```

```python
diags_new_id = pd.merge(diags_copy, new_diags, on='diagnostic', how='left')
diags_new_id = diags_new_id.drop(["cancer", "id_label", "generation", 'statut'], axis=1)
diags_new_id = diags_new_id.rename(columns={'id_diag_x': 'id_diag'})
diags_new_id.head(5)
```

```python
diags_new_id.id_diag = diags_new_id.id_diag.astype(int)
diags_new_id.id_diag_y = diags_new_id.id_diag_y.astype(int)
CRO_3_transco = pd.merge(CRO_3, diags_new_id[['id_diag', 'id_diag_y']], on='id_diag', how='left')
CRO_3_transco.head(5)
```

```python
CRO_3_transco = CRO_3_transco.drop(['id_diag'], axis=1)
CRO_3_transco = CRO_3_transco.rename(columns={'id_diag_y': 'id_diag'})
CRO_3_transco.head(5)
```

```python
CRO_1_et_2_et_3 = pd.concat([CRO_1_et_2, CRO_3_transco], ignore_index=True )
CRO_1_et_2_et_3.head(5)
```

```python
CRO_1_et_2_et_3.to_csv(f'{file}/CRO_1_2_3.csv', index=False, sep=';')
```

```python
diags_v3 = pd.read_csv('../data/diagnostics.csv', sep=';', dtype=str)
diags_new_id_bis = pd.merge(diags_v3, new_diags, on='diagnostic', how='left')
diags_new_id_bis = diags_new_id_bis.drop(["cancer", "generation", 'statut'], axis=1)
diags_new_id_bis = diags_new_id_bis.rename(columns={'id_diag_x': 'id_diag'})
diags_new_id_bis.head(5)
```

```python
diags_new_id_bis = diags_new_id_bis[['id_diag', 'id_diag_y']]
diags_new_id_bis = diags_new_id_bis.astype(int)
CRO_4_transco = pd.merge(CRO_4, diags_new_id_bis, on='id_diag', how='left')
CRO_4_transco = CRO_4_transco.drop(['id_diag'], axis=1)
CRO_4_transco = CRO_4_transco.rename(columns={"id_diag_y": "id_diag"})
CRO_4_transco.head(5)
```

```python
CRO_1_et_2_et_3_et_4 = pd.concat([CRO_1_et_2_et_3, CRO_4], ignore_index=True)
CRO_1_et_2_et_3_et_4.to_csv(f'{file}/CRO_1_2_3_4.csv', index=False, sep=';')
```

```python
CRO = pd.concat([CRO_1_et_2_et_3_et_4, CRO_5], ignore_index=True)
CRO.to_csv(f'{file}/CRO.csv', index=False, sep=';')
```
